```python
import matplotlib.pyplot as plt
```

```python
def fx1(v):
    return (v*v)
def fx2(v):
    return ((v**4)-5*v*v-3*v)
def grad1(v0):
    return 2*v0
def grad2(v0):
    return (4*(v0**3)-10*v0-3)
```

```python
epochs=1000

x=list(range(epochs))
y=[]
y1=[]
def gradient(v0,eta,flag):
    i=0
    while i<epochs :
        if flag==1:
            v_grad=grad1(v0)
        if flag==2:
            v_grad=grad2(v0)
        if v_grad==0:
            return 
        v0=v0-eta*v_grad
        y1.append(v0)
        print(v0)
        if flag==1:
            print("function value",fx1(v0))
            y.append(fx1(v0))
            
        if flag==2:
            print("function value",fx2(v0))
            y.append(fx2(v0))
        
        i=i+1
```

```python
gradient(10,0.1,1)
```

```python
plt.scatter(x, y,color='red')
plt.xlabel("Iterations")
plt.ylabel("Value of C")
plt.show()
```

```python
plt.scatter(x, y1,color='red')
plt.xlabel("Iterations")
plt.ylabel("Value of v")
plt.show()
```
